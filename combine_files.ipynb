```python
import pandas as pd
import glob
```

```python
data_dir = "/Users/katybarone/Documents/projects/tennis/"

atp_year_lst = glob.glob(data_dir + "atp*.csv")

matches_dfs = []
for year in atp_year_lst:
    df_temp = pd.read_csv(year)
    
    matches_dfs.append(df_temp)

df_matches = pd.concat(matches_dfs, ignore_index=True)

df_matches.head()
```

```python
atp_100 = pd.read_csv(data_dir + "top_100.csv")
print(len(atp_100))

# Auger-Aliassime and Jan-Lennard need to not have dashes, O'connect can't have an apostrophe
atp_100['name'] = atp_100['name'].str.replace("-", " ")
atp_100['name'] = atp_100['name'].str.replace("'", "")

print(atp_100[atp_100['name'].str.contains("Felix")])
print(atp_100[atp_100['name'].str.contains("Lennard")])
print(atp_100[atp_100['name'].str.contains("Christopher")])

atp_100 = atp_100.merge(df_matches[["loser_name", "loser_id"]], left_on="name", right_on="loser_name", how="inner")

atp_100 = atp_100.drop_duplicates()
atp_100 = atp_100.drop(columns="loser_name", axis=1)
atp_100 = atp_100.rename(columns={"loser_id":"id"})

print(len(atp_100))
```

```python
atp_100.head()
```

```python
atp_100.to_csv("atp_100_withIDs.csv", index=False)
```

```python
top_ids = atp_100['id'].values

df_100_matches = df_matches[(df_matches['loser_id'].isin(top_ids)) & (df_matches['winner_id'].isin(top_ids))]
df_100_matches.head()
```
